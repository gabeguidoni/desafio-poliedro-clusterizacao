```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
df_escolas = pd.read_csv("../dados/temporarios/02-dados_escolas.csv",
                         usecols=["CO_ENTIDADE", "lat", "lon", "valor_venda"])
df_escolas.head()
```

```python
df_consultores = pd.read_csv("../dados/temporarios/01-consultor_coords.csv")

df_consultores.head()
```

```python


fig = go.Figure()


fig.add_trace(go.Scattermap(
    lat=df_escolas["lat"],
    lon=df_escolas["lon"],
    mode="markers",
    marker=dict(
        color = pd.qcut(df_escolas["valor_venda"], q=10, labels=False),
        colorscale="Viridis",
        reversescale = True,
        showscale=True,
    ),
    text=round(df_escolas["valor_venda"]/1000),
    name="Escolas",
))


fig.add_trace(go.Scattermap(
    lat=df_consultores["lat"],
    lon=df_consultores["lon"],
    mode="markers",
    marker=dict(
        size=10,
        color="red"
    ),
    text=df_consultores["Consultor"],
    name="Consultores"
))

fig.update_layout(
    map=dict(
        style="open-street-map",
        center=dict(lat=-14.2350, lon=-51.9253),
        zoom=3
    ),
    legend=dict(
        x=0.02,
        y=0.98,
        xanchor="left",
        yanchor="top"
    ),
    height=700,
    margin=dict(r=0, t=0, l=0, b=0),
    

)

fig.show()
```
