```python
df = df_raw[df_raw["TP_DEPENDENCIA"] == 4]  # Mantem apenas escolas particulares, reduz para 52_568 escolas
df = df[(df["TP_CATEGORIA_ESCOLA_PRIVADA"] != 3)]  # Remove escolas particulares confessionais, reduz para 52_101 escolas
df = df[(df["TP_SITUACAO_FUNCIONAMENTO"] == 1)]  # Mantem apenas escolas em atividade, reduz para 42_751 escolas

df = df.drop(columns=["TP_DEPENDENCIA", "TP_CATEGORIA_ESCOLA_PRIVADA", "TP_SITUACAO_FUNCIONAMENTO"])

df.info()
```
