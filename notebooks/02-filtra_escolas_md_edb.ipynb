- filtrar somente as escolas de interesse (particular, nao confessiona e em atividade)
- remover escolas com menos de 20 alunos no total (pegar as associadas e confirmar esse numero minimo)

ALERTA: Algumas escolas nao possuem nenhum numero de alunos preenchido, e entao foram eliminadas. (Avisar)

```python
import pandas as pd


pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_seq_items', None)
```

```python
df_raw = pd.read_csv("../dados/temporarios/01-md_edb_colunas.csv")

df_raw.shape

# 215 mil escolas
```

```python
df_raw.info()
```

```python
df = df_raw[df_raw["TP_DEPENDENCIA"] == 4]  # Mantem escolas particulares, reduz para 52_568
df = df[(df["TP_CATEGORIA_ESCOLA_PRIVADA"] != 3)]  # Remove escolas particulares confessionais, reduz para 52_101
df = df[(df["TP_SITUACAO_FUNCIONAMENTO"] == 1)]  # Mantem escolas em atividade, reduz para 42_751

df = df.drop(columns=["TP_DEPENDENCIA", "TP_CATEGORIA_ESCOLA_PRIVADA", "TP_SITUACAO_FUNCIONAMENTO"])
```

```python
df.head()
```

```python
df["QT_TOTAL_MAT"] = (
      df["QT_MAT_INF"].fillna(0)
    + df["QT_MAT_FUND_AI"].fillna(0)
    + df["QT_MAT_FUND_AF"].fillna(0)
    + df["QT_MAT_MED"].fillna(0)
    )

df = df[df["QT_TOTAL_MAT"] > 20] # Mantem escolas com mais de 20 alunos, reduz para 37_183
```

```python
df.to_csv("../dados/temporarios/02-md_edb_limpo.csv", index=False)
```
